# As we can see, company and agent has too many null values, we'll delete them

```python
hotels = hotels.drop(['company', 'agent'], axis = 1)
```

```python
hotels
```

# The columns country and children have a small number of missing values.But they can be important factor for the analysis, so here we are just removing those rows where we have null values in these columns.

```python
hotels[hotels['country'].isnull()]
```

```python
hotels[hotels['children'].isnull()]
```

```python
hotels=hotels.dropna(axis=0).reset_index(drop=True)
```

```python
hotels.head()
```

```python
hotels.isna().sum().sort_values(ascending = False)
```

```python
#Checking where there is no guests.
hotels[hotels.adults + hotels.babies + hotels.children == 0].shape
```

# These rows does not make any sense, so we are gonna delete these rows

```python
hotels = hotels.drop(hotels[(hotels.babies + hotels.children + hotels.adults == 0)].index).reset_index(drop = True)
```

```python
hotels
```
